```python
import pandas as pd
import numpy as np
import seaborn as sns


# Import the libraries we'll use to create graphs
from matplotlib import pyplot as plt

# This command makes sure the graphs show in this window, not a new one
%matplotlib inline 
```

```python
# import the csv file as a dataframe
url = 'https://raw.githubusercontent.com/CS222-UIUC/course-project-group-102/main/Training_Dataset.csv'
data = pd.read_csv(url)
```

```python
count_row=data.shape[0]
print(count_row) #displaying the number of rows (datapoints)
```

```python
count_column=data.shape[1]
print (count_column) #number of features
```

```python
print(data.columns)
```

```python
data.head()
```

```python
#made a dataframe out of data variable
df = pd.DataFrame(data)
#conversion of all numeric values to float
df['AVERAGE STOCK PRICE'] = df['AVERAGE STOCK PRICE'].str.replace(',', '')
df['AVERAGE STOCK PRICE'] = df['AVERAGE STOCK PRICE'].astype('float')
print (df)
print (df.dtypes)
```

```python
# distrubtion plot showing the correlation between stock price and house price

sns.displot(df, x="AVERAGE STOCK PRICE", y="AVERAGE HOUSE PRICE", legend=True).set(title='Stock Prices Distribution Vs. Housing Prices Distribution')
```

```python
sns.set_theme(style="darkgrid")

# Plot the responses for different events and regions
sns.lineplot(x="AVERAGE STOCK PRICE", y="AVERAGE HOUSE PRICE", data=df)
```

```python

sns.jointplot(x="AVERAGE STOCK PRICE", y="AVERAGE HOUSE PRICE", data=df,
                  kind="reg", truncate=False,
                  color="m", height=7)
```
